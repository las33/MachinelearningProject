```python
import matplotlib.pyplot as plt

def create_plot_fp_simple(basename, basename_styled, base_index, alg1, alg1_name,alg2, alg2_name,max_m, max_time,secCor='r'):
    

    memory_alg1 = [i['memory'] for x, i in alg1[base_index]["databases/"+basename+".txt"].items()]
    memory_alg2 = [i['memory'] for x, i in alg2[base_index]["databases/"+basename+".txt"].items()]

    time_alg1 = [i['time'] for x, i in alg1[base_index]["databases/"+basename+".txt"].items()]
    time_alg2 = [i['time'] for x, i in alg2[base_index]["databases/"+basename+".txt"].items()]

    suports = [x for x in alg2[base_index]["databases/"+basename+".txt"]]  
    
    
    
    if(len(suports) != len(memory_alg1)):
        plt.plot(suports[:-1],memory_alg1 , color=secCor, label=alg1_name) 
    else:
        plt.plot(suports, memory_alg1, color=secCor, label=alg1_name)  
    plt.plot(suports,memory_alg2 , color='b', label=alg2_name)  
    plt.xlabel('Suporte minimo %')
    plt.ylabel('Consumo de memoria MB')
    plt.title(basename_styled)
    
    interval = int(max_m / 9) 
    plt.yticks(np.arange(0, max_m + 1, interval)) 
    plt.xticks(suports) 

    plt.legend()
    plt.show()    
      
    if(len(suports) != len(time_alg1)):
        plt.plot(suports[:-1],time_alg1 , color=secCor, label=alg1_name) 
    else:
        plt.plot(suports, time_alg1, color=secCor, label=alg1_name)  
        
    plt.plot(suports,time_alg2 , color='b', label=alg2_name) 
    plt.xlabel('Suporte minimo %')
    plt.ylabel('Tempo de Processamento s')
    plt.title(basename_styled)
    interval = int(max_time / 9) 
    plt.yticks(np.arange(0, max_time+1, interval)) 
    plt.xticks(suports) 

    plt.legend()
    plt.show()
```

```python
create_plot_fp_simple('mushroom', "Mushroom",0, results_fp, "FP+",results_simple,"PrePost+",900, 90)
```

```python
create_plot_fp_simple('connect', "Connect",1, results_modified, "PrePost+ Modificado",results_simple,"PrePost+ Simples",900, 250, 'g')
```
